```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(['a', 'b', 'c', 'd', 'e'])
y = np.array([24, 85, 47, 38, 62])
```

```python
# Plot vertical bar chart
plt.bar(x, y, color = 'y')
```

```python
# plot horizontal bar chart
plt.barh(x, y)
```

```python
# Add colors to the bar
plt.barh(x, y, color='g')
plt.title("Horizotal Bar graph")
```

```python
# plot bar chart with specified width
plt.bar(x, y, color='orange', width=0.2)
```

```python
# Plot the graph with specific height
plt.barh(x, y, color ='y', height = 0.2)
```

```python
# Create pie chart
plt.pie(y, labels=x)
plt.title("Pie chart")
plt.show()
```

```python
# Create pie chart with exploded view
plt.pie(y, labels=x, explode=[0.2, 0, 0, 0, 0])
plt.show()
```
