```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]
```

```python
# Plot line graph
plt.plot(x, y)
plt.title('Line Graph')
plt.show()
```

```python
# Create 100 values between 0 and 10
x1 = np.linspace(0, 10, 100)
x1
# fig=plt.figure()
```

```python
# plot sine wave
plt.plot(x1, np.sin(x1), '-')
plt.show()
```

```python
# plot cosine wave
plt.plot(x1, np.cos(x1), '--')
plt.show()
```

```python
# Create subplot and plot sin and cosine wave individually
plt.subplot(2,2,1)
plt.plot(x1, np.sin(x1))
plt.subplot(2,2,2)
plt.plot(x1, np.cos(x1), '--')
```

```python
# Plot is a very versatile command 
plt.plot([1, 2, 3, 4], [1, 4, 8, 16])
plt.show()
```
