```python
# groupby will group the features and will do the neccessary operations
df2 = df1[['Geography', 'Gender', 'Balance']].groupby(['Geography', 'Gender'])
df2.head()
```

```python
# Applying mean operations in the balance column for geography and gender
df2 = df1[['Geography', 'Gender', 'Balance']].groupby(['Geography', 'Gender']).mean()
df2.head()
```

```python
df1[['Geography', 'Gender', 'Balance']].groupby(['Geography', 'Gender']).agg(['mean', 'count', 'sum'])
```
