```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Science':[85, 65, 72, 47, 53],
                   'Maths':[56, 73, 66, 92, 68],
                   'English':[54, 88, 49, 82, 75]})
df1
```

```python
df2 = pd.DataFrame({'Science':[91, 48],
                   'Maths':[75, 82],
                   'English':[64, 95]})
df2
```

```python
df3_melted = df1.melt().rename(columns={'variable': 'Subject', 'value': 'Marks'})
df3_melted
```

```python
# Pivot rows as columns
df3_melted.pivot(columns='Subject', values='Marks')
```

```python
# Concatenate 2 dataframes
pd.concat([df1, df2])
```

```python
df4 = pd.DataFrame({'Hindi':[43, 32],
                   'Biology':[67, 76],
                   'Chemistry':[86, 76]})
df4
```

```python
# Add two dataframes horizontally(as columns)
pd.concat([df2, df4], axis=1)
```

```python
df3_melted
```

```python
# By default the values are sorted in ascending order
df3_melted.sort_values(by='Marks')
```

```python
# ascending = False will sort the values in descending order
df3_melted.sort_values(by='Marks', ascending=False)
```

```python
# Find MIN and MAX of columns
science_min = df1['Science'].min()
science_max = df1['Science'].max()
print(science_min)
print(science_max)
```

```python
index_values_1 = ['A', 'B', 'C', 'D', 'E']
index_values_2 = ['A', 'B']
df1.insert(0, 'Names', index_values_1)
df4.insert(0, 'Names', index_values_2)

df1.set_index('Names', inplace=True)
df4.set_index('Names', inplace=True)
df5 = pd.merge(df1, df4, how='left', on='Names')
print(df1)
print(df4)
print(df5)
```
