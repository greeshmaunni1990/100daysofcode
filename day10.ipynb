```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = np.random.randn(100)
plt.hist(data1)
plt.show()
```

```python
# Bar chart
data2 = [5. , 25. , 50. , 20.]
print(range(len(data2)))
plt.bar(range(len(data2)), data2)
```

```python
# Horizontal bar chart
plt.barh(range(len(data2)), data2)
```

```python
# Stacked bar chart
a = [15., 30., 45., 22.]
b = [15., 25., 50., 20.]
z1 = range(4)
plt.bar(z1, a, color='b')
plt.bar(z1, b, color='g')
plt.title("Stacked Bar")
plt.show()
```

```python
# Pie chart
x1 = [35, 25, 20, 20]
labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']
plt.pie(x1, labels=labels)
plt.title("Pie Chart")
plt.show()
```

```python
# Box Plot
data3 = np.random.randn(100)
plt.boxplot(data3)
plt.title("Box PLot")
plt.show()
```

```python
# Area Chart
x2=range(1,6)
y2=[1, 4, 6, 8, 4]
plt.fill_between(x2, y2)
plt.title('Area Chart')
plt.show()
```

```python
# Contour plot
matrix1 = np.random.rand(10,20)
plt.contour(matrix1)
plt.title("Contour plot")
plt.show()
```
