```python
import pandas as pd
import numpy as np
```

```python
pd_dict = {'Gender':['Male',np.nan, 'Female', np.nan, 'Female'],
          'Science':[85, np.nan, 72, 47, 53],
          'Maths':[56, np.nan, np.nan, 92, 68],
          'English':[np.nan, np.nan, 49, 82, 75]}
```

```python
df1 = pd.DataFrame(pd_dict)
df1.head()
```

```python
# Check if null value exists
df1.isnull()
```

```python
# Check the opposite of isnull()
df1.notnull()
```

```python
# Fill null values with 1
df1.fillna("1")
```

```python
# Fill the null values with previous values
df1.fillna(method='ffill')
```

```python
# Fill the null values with next values
df1.fillna(method='bfill')
```

```python
# Filling null values with Not specified
df1['Gender'].fillna('Not Specified', inplace=True)
df1
```

```python
# Replace the null value witg=h 1
df1[['Science']].replace(to_replace=np.nan, value=1)
```

```python
# Interpolate(insert) missing values using linear method
df1.interpolate(method='linear', limit_direction='forward')
```
